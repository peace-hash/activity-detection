```python
import numpy as np 
import pandas as pd 

data_paths = {}
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.pkl'):
            data_paths[filename] = os.path.join(dirname, filename)
        
data_paths
```

```python
data = {}
for file in data_paths:
    data[file] = pd.read_pickle(data_paths[file])
```

```python
data['train_data.pkl']['sensor'].value_counts()
```

```python
data['train_data.pkl']['device'].value_counts()
```

```python
sensor_map ={'accel':1, 'gyro':2}
device_map ={'phone':1, 'watch':2}

for file in data_paths:
    data[file]['sensor'] = data[file]['sensor'].replace(sensor_map)
    data[file]['device'] = data[file]['device'].replace(device_map)
```

```python
ys = {}
for file in data_paths:
    ys[file] = data[file]['target']


drop_cols = ['target','idx']


Xs = {}
for file in data_paths:
    Xs[file] = data[file].drop(drop_cols, axis = 1)
```

```python
Xs['train_data.pkl'].shape, Xs['test_data.pkl'].shape, Xs['val_data.pkl'].shape
```

```python
ys['train_data.pkl'].shape, ys['test_data.pkl'].shape, ys['val_data.pkl'].shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

model = RandomForestClassifier(n_jobs = -1, class_weight = 'balanced', n_estimators=800)
model.fit(Xs['train_data.pkl'], ys['train_data.pkl'])
```

```python
test_predicts = model.predict(Xs['test_data.pkl'])
print(metrics.accuracy_score(ys['test_data.pkl'], test_predicts))
```

```python
val_predicts = model.predict(Xs['val_data.pkl'])
print(metrics.accuracy_score(ys['val_data.pkl'], val_predicts))
```
